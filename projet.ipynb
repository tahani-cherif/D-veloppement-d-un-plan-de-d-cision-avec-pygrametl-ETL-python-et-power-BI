```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h1>1.Ouvrir le fichier et convertir la date en clé primaire </h1>

```python
data=pd.read_csv('metacritic_video_games.csv',index_col='date',parse_dates=True)
data = data.dropna(axis=0)
data.head()
```

<h1>2.Taille de base de donnée  </h1>

```python
data.shape
```

<h1>3.Présenter le nombre de jeux pour chaque genre    </h1>

```python
font1 = {'family':'serif','color':'blue','size':19}
plt.title("Nombre de jeux de chaque genre",fontdict = font1)
plt.xlabel("Genre",fontdict={'size':20})
plt.ylabel("Nombre de jeux",fontdict={'size':17})
color={"lightcoral","maroon","mistyrose","coral","peachpuff","darkorange","navajowhite","orange","darkgoldenrod","gold","ivory","olive","yellowgreen","lawngreen","lightgreen","g","mediumseagreen","mediumaquamarine"}
data['genre'].value_counts().plot.bar(color=color,figsize=(12,3.5))
```

<h1>4.présenter chaque part du type de platforme Xbox </h1>

```python
platforms=data.groupby(['platforms']).count()
```

```python
platforms= platforms.drop(['3DS','DS','Dreamcast','Game Boy Advance','GameCube','Nintendo 64','PC','PSP','Stadia','Switch','Wii','Wii U','PlayStation','PlayStation 2','PlayStation 3','PlayStation 4','PlayStation 5','PlayStation Vita','iOS'])
platforms=platforms[:-1]
part=(platforms['titles']/sum(platforms['titles']))
print(part)
label='Xbox','Xbox360','Xboxone','XboxSeriesX'
```

```python
explode =(0.1, 0.1 , 0.1 , 0.1 )
plt.title("Pourcentage chaque type de Xbox")
plt.pie(part,explode=explode,autopct='%.2f%%')
plt.legend(label,title="legend", loc="center left",bbox_to_anchor=(1.5,0, 0, 1))
cer= plt.gcf()
cer.set_size_inches(5,5)
```

<h1>5.Présenter nombre de jeux de chaque platforme crée en 2020 </h1>

```python
plt.title("nombres de jeux de chaque platform en 2020",fontdict={'size':20 , 'color':'blue'})
plt.xlabel("platform",fontdict={'size':20})
plt.ylabel("Nombre de jeux",fontdict={'size':17})
data.loc['2020','platforms'].value_counts().plot(figsize=(11,5))
```

<h1>6.Présenter le metascore maximum pour chaque genre</h1>

```python
data.groupby(['genre'])['metascore'].max()
```

```python
plt.title("Le maximum Metascore de chaque genre",fontdict={'size':20 , 'color':'blue'})
plt.xlabel("Genre",fontdict={'size':20})
plt.ylabel("Metascore",fontdict={'size':17})
data.groupby(['genre'])['metascore'].max().plot(figsize=(11,5))
```
